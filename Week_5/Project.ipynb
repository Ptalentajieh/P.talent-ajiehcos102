```python
import os
import csv
import tkinter as tk
from tkinter import messagebox

# Get the path to the CSV file
csv_file_path = os.path.join(os.path.dirname(__file__), "Gig.csv")

# Load employee data from the CSV file
def load_employee_data(filename):
    employee_list = []
    with open(filename, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            employee_list.append(row)
    return employee_list

# Function to check if the employee exists
def verify_employee():
    entered_name = name_input.get().strip().lower()
    entered_dept = dept_input.get().strip().lower()

    found = False
    department_members = []

    for employee in employee_data:
        employee_name = employee['Name'].strip().lower()
        employee_dept = employee['Department'].strip().lower()

        if employee_name == entered_name and employee_dept == entered_dept:
            found = True
        elif employee_dept == entered_dept:
            department_members.append(employee['Name'])

    if found:
        message = f"Welcome, {name_input.get().strip()} from {dept_input.get().strip()} department!"
        if department_members:
            message += "\n\nOther members in your department:\n- " + "\n- ".join(department_members)
        else:
            message += "\n\nYou are the only one in your department."
        messagebox.showinfo("Employee Found", message)
    else:
        messagebox.showwarning("Not Found", f"Sorry, {name_input.get().strip()} from {dept_input.get().strip()} department is not in our records.")

# Load data from CSV
employee_data = load_employee_data(csv_file_path)

# Create the GUI window
window = tk.Tk()
window.title("GIG Logistics - Employee Checker")

# Add labels and input fields (no padding used)
tk.Label(window, text="Enter Name:").grid(row=0, column=0, sticky='e')
name_input = tk.Entry(window, width=30)
name_input.grid(row=0, column=1)

tk.Label(window, text="Enter Department:").grid(row=1, column=0, sticky='e')
dept_input = tk.Entry(window, width=30)
dept_input.grid(row=1, column=1)

# Add verify button
verify_button = tk.Button(window, text="Verify Employee", command=verify_employee)
verify_button.grid(row=2, column=0, columnspan=2)

# Run the application
window.mainloop()
```
